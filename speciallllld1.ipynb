```python
import pandas as pd

df1 = pd.read_excel("/content/dataset1.xlsx")
df2 = pd.read_excel("/content/dataset2.xlsx")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for df, name in [(df1, "Dataset 1"), (df2, "Dataset 2")]:
    corr = df.iloc[:, 2:].corr()
    plt.figure()
    sns.heatmap(corr, annot=True)
    plt.title(f"Correlation Matrix - {name}")
    plt.show()
```

```python
for df, name in [(df1, "Dataset 1"), (df2, "Dataset 2")]:
    df_dev = df.copy()

    for col in df.columns[2:]:
        df_dev[col] = df[col] - df["Age"]

    plt.figure()
    sns.heatmap(df_dev.iloc[:, 2:])
    plt.title(f"Age Deviation - {name}")
    plt.show()
```

```python
for df, name in [(df1, "Dataset 1"), (df2, "Dataset 2")]:
    plt.figure()

    for col in df.columns[2:]:
        plt.scatter(df["Age"], df[col], label=col)

    plt.xlabel("Chronological Age")
    plt.ylabel("Predicted Age")
    plt.title(f"Predicted vs Actual Age - {name}")
    plt.legend()
    plt.show()
```
